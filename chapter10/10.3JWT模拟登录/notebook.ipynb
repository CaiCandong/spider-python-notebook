```python
from wsgiref import headers
import requests
url  = "https://login3.scrape.center/api/login"

data = {
    "username":'admin',
    "password":'admin',
}
# 禁止自动重定向
rep= requests.post(url,json = data)
# requests 自动解析headers中的set-cookies
# rep.status_code,rep.json()
jwt = rep.json().get('token')
headerS = {
    "Authorization":f'jwt {jwt}'
}
rep= requests.get("https://login3.scrape.center/api/book",params={'limit':18,'offset':0},headers=headerS)
rep.json()
```
