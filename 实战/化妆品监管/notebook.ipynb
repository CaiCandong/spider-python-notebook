```python
from selenium import webdriver

from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
def get_browser():
        
    # chrome 浏览器位置
    # 驱动参数配置
    options = Options()
    options.add_experimental_option('excludeSwitches',['enable-automation'])
    options.add_experimental_option('useAutomationExtension',False)

    # option.add_argument('headless')  # 后台运行
    options.binary_location = r"C:\Program Files\Google\Chrome\Application\chrome.exe"

    # chrome 浏览器驱动位置
    chrome_service = Service(r'C:\Users\root\anaconda3\Scripts\chromedriver.exe')


    browser = webdriver.Chrome(
        options=options, 
        service =chrome_service)


    browser.execute_cdp_cmd('Page.addScriptToEvaluateOnNewDocument',{
        'source':'Object.defineProperty(navigator,"webdriver",{get:()=>undefined})'
    })
    return browser
```

```python
browser=get_browser()
url ="http://scxk.nmpa.gov.cn:81/xk/"
browser.get(url)
```

```python

from selenium.webdriver.common.by import By
def parse_index():
    conpanys =browser.find_elements(by=By.TAG_NAME,value="li")[1:]
    infos = []
    for company in conpanys:
        # 发证日期
        date = company.find_element(by=By.TAG_NAME,value="i").text
        # 公司名字
        name = company.find_element(by=By.TAG_NAME,value="dl").text
        # 超连接
        hyper=company.find_element(by=By.TAG_NAME,value="a").get_attribute("href")
        # 发证编号
        no=company.find_element(by=By.TAG_NAME,value="ol").text
        # 发证机关
        issuer= company.find_element(by=By.TAG_NAME,value="p").text
        # 有效期
        deadtate = company.find_element(by=By.TAG_NAME,value="em").text
        infos.append({
            '发证日期':date,
            '公司名字':name,
            '超连接':hyper,
            '发证编号':no ,
            '发证机关':issuer,
            '有效期':deadtate
        })
    return infos
parse_index()
```

```python
import time
browser= get_browser()
url ="http://scxk.nmpa.gov.cn:81/xk/"
browser.get(url)
time.sleep(1)
pageIto_next=browser.find_element(by=By.ID,value='pageIto_next')
pageIto_next.click()
```
